# Setup LLM

```python
import ollama
import gc
import torch

# To run others llms, just change the model name and id below and run this cell again to clear the cache and load the new model.
model_name="Deepseek"
MODEL_ID="deepseek-r1:7b"

for var in ['model', 'pipe', 'tokenizer']:
    if var in globals():
        del globals()[var]

gc.collect()
torch.cuda.empty_cache()
```

## Function to send the post and the prompt for LLM analysis

```python
def run_batch_analysis(dataset_obj, prompt_template):
    results = []
    total = len(dataset_obj)
    
    for i, row in enumerate(dataset_obj):
        print(f"Processing {i+1}/{total}...", end="\r")
        text = str(row["Text"])
        
        # Replace the placeholder in your original prompt
        prompt = prompt_template.replace("{excerpt}", text)
        
    
        response = ollama.chat(
                model=MODEL_ID,
                messages=[{'role': 'user', 'content': prompt}],
                stream=False,
                
            )
        # Extract the generated text from the response
        generated_text = response['message']['content'].strip()
        results.append(generated_text)
        
    # Return the list of results
    return results
```

# Classification

```python
prompts_subtemas = [fewshot,oneshot,zeroshot]
prompts_polarizacao = [fewshot_polarization,oneshot_polarization,zeroshot_polarization]
prompt_names = ['fewshot','oneshot','zeroshot']
```

```python
from pydantic import BaseModel, Field
from typing import Literal

# --- Schemes for Subthemes ---
class SubthemeValues(BaseModel):
    gender: Literal[0, 1]
    race: Literal[0, 1]
    lgbtq_plus: Literal[0, 1] = Field(..., alias="lgbtq+")
    disability: Literal[0, 1]
    none: Literal[0, 1]

class SubthemeSchema(BaseModel):
    subtheme: SubthemeValues
    justification: str

# --- Schemes for Polarization ---
class PolarizationSchema(BaseModel):
    polarization: Literal["pro-DEIA", "anti-DEIA", "neutral"]
    justification: str
```

```python
output_file = f"PANDEIA_RESULTS_{model_name}.csv"

# Load existing results if the output file already exists, otherwise start with the original dataset
if os.path.exists(output_file):
    print(f"Starting from existing file: {output_file}")
    df = pd.read_csv(output_file)
else:
    print("Starting new classification from PANDEIA_DATASET.csv")
    df = pd.read_csv('PANDEIA_DATASET.csv')



# Loop for each prompt type (fewshot, oneshot, zeroshot) and perform classification
for i in range(len(prompt_names)):
    name = prompt_names[i]
    col_subtheme = f'subthemes_{name}'
    col_polarization = f'polarization_{name}'
    
    if col_subtheme not in df.columns:
        df[col_subtheme] = None
    if col_polarization not in df.columns:
        df[col_polarization] = None

    print(f"\n--- Starting technique: {name} ---")
    
    # Iterate through each row in the DataFrame and classify subtheme and polarization
    total_rows = len(df)
    for idx, row in df.iterrows():
        text = str(row["Text"])
        
        # 1. Classification of Subthemes (if not already filled)
        if pd.isna(df.at[idx, col_subtheme]) or df.at[idx, col_subtheme] == "":
            print(f"[{name}] Classifying Subtheme - Row {idx+1}/{total_rows}...", end="\r")
            prompt_sub = prompts_subtemas[i].replace("{excerpt}", text)
            
            try:
                resp_sub = ollama.chat(
                    model=MODEL_ID,
                    messages=[{'role': 'user', 'content': prompt_sub}],
                    format=SubthemeSchema.model_json_schema(),
                    stream=False
                )
                df.at[idx, col_subtheme] = resp_sub['message']['content'].strip()
                
                # Saving immediately after classifying subtheme
                df.to_csv(output_file, index=False)
            except Exception as e:
                print(f"\nError on row {idx} (Subtheme): {e}")

        # 2. Classification of Polarization (if not already filled)
        if pd.isna(df.at[idx, col_polarization]) or df.at[idx, col_polarization] == "":
            print(f"[{name}] Classifying Polarization - Row {idx+1}/{total_rows}...", end="\r")
            prompt_pol = prompts_polarizacao[i].replace("{excerpt}", text)
            
            try:
                resp_pol = ollama.chat(
                    model=MODEL_ID,
                    messages=[{'role': 'user', 'content': prompt_pol}],
                    format=PolarizationSchema.model_json_schema(),
                    stream=False
                )
                df.at[idx, col_polarization] = resp_pol['message']['content'].strip()
                
                # Saving immediately after classifying polarization
                df.to_csv(output_file, index=False)
            except Exception as e:
                print(f"\nError on row {idx} (Polarization): {e}")

print(f"\nProcess completed! Final file saved to: {output_file}")
```
